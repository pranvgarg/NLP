```python
#12 BASIC TEXT PROCESSING PIPELINE
import nltk
text1='''Originally, I was disappointed by the high prices of these watches. I didn't want to drop $700 on a watch when I wasn't sure I would even use a watch. However, I ended up being very happy with the Sport, and I'm convinced it's what I would have chosen even if my funds allowed me other choices-- at least, that's what I tell myself. :) My mom, who also likes to have the latest gadget, got the Watch version, meaning I have a pretty good basis of comparison, so I'll start with that. '''
for text in text1:
    sentences=nltk.sent_tokenize(text1)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        tagged=nltk.pos_tag(words)
        print(tagged)
```

```python
#16 SIMPLE TEXT CLASSIFICATION
def gender_features(word):
    return{'last_letter':word[-1]}
```

```python
from nltk.corpus import names
```

```python
names.words()
```

```python
labeled_names=([(name,'male')for name in names.words('male.txt')]+[(name,'female')for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)
featuresets=[(gender_features(n),gender)for (n,gender) in labeled_names]
```

```python
train_set,test_set=featuresets[:5000],featuresets[5000:]
import nltk
classifier=nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_features('David'))
```

```python
classifier.classify(gender_features('Pranav'))
```

```python
classifier.classify(gender_features('Pulkit'))
```

```python
print(nltk.classify.accuracy(classifier,test_set))
```

```python
#17 COUNT VECTORIZER
from sklearn.feature_extraction.text import CountVectorizer
vect=CountVectorizer(binary=True)
corpus=["Tessaract is good optical character recognition engine ","optical character recognition is significant"]
vect.fit(corpus)
```

```python
vocab=vect.vocabulary_
```

```python
for key in sorted(vocab.keys()):
    print("{}:{}".format(key,vocab[key]))
```

```python
print(vect.transform(["This is a good optical illusion"]).toarray())
```

```python
#18 FINDING SIMILARITY BETWEEN DOCUMENTS
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vect.transform(["Google Cloud Vision is a character recognition engine"]).toarray(),vect.transform(["OCR is an optical character recognition engine"]).toarray())
print(similarity)
```
